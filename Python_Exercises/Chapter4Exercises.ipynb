Rewrite your pay computation with time-and-a-half for overtime and create a function called compute pay which takes two parameters (hours and rate).

Enter Hours: 45

```python
def compute_pay(hours, rate):
    if hours <= 40:
        pay = hours * rate
    else:
        reg_pay = 40 * rate
        overt_pay = (hours - 40) * (rate * 1.5)
        pay = reg_pay + overt_pay
    return pay


hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))
print("Pay:", compute_pay(hours, rate))
```

**Challenge**
Write a function called computegrade that takes a score as its parameter and returns a grade as a string using the following table:

Score Grade [ 97, 100] 1.0 [ 93, 97) 1.25 [ 89, 93) 1.5 [ 85, 89) 1.75 [ 81, 85) 2.0 [ 77, 81) 2.25 [ 74, 77) 2.5 [ 0, 74) 3.0

```python
def computegrade(score):
    if score >= 97:
        return '1.0'
    elif score >= 93:
        return '1.25'
    elif score >= 89:
        return '1.5'
    elif score >= 85:
        return '1.75'
    elif score >= 81:
        return '2.0'
    elif score >= 77:
        return '2.25'
    elif score >= 74:
        return '2.5'
    elif score >= 0:
        return '3.0'


score = float(input("Enter the score: "))
print("Grade:", computegrade(score))
```
